```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = the more education that you have the higher your salary will be

```python
# calculate correlation and pvalue of salary and education with spearmanr method
correlation_coefficient, p_value = spearmanr(df['salary'], df['education'])
print(f"Spearman correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
```

with the spearman correlation being about 0 it means there is a moderate positive correlation

#### Are these results statistically significant? How do you know?

yes, because the p-value is less than 0.05 meaning it is unlikely the have happen by chance

### What is the correlation between salary and age?

#### Your Hypothesis: the more older you are the more money you make

```python
# calculate correlation and pvalue of salary and age with pearsonr method
correlation_coefficient, p_value = pearsonr(df['salary'], df['age'])
print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
```

the pearson correlation coefficient is again above 0 and under 1 meaning they likely correlate

#### Are these results statistically significant? How do you know? 

yes these are significant with the p-value of 0 because it shows that they correlate to each other

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = the longer you are in the field the more money you will make

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
correlation_coefficient, p_value = pearsonr(df['salary'], df['industry_experience'])
print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
```

same as the above two these two correlate and effect each other

#### Are these results statistically significant? How do you know?

yes it is significant because it means they affect each other

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(df['education'], df['industry'], values=df['salary'], aggfunc='mean', margins=True, margins_name='Average')
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
```

```python
average_salary_crosstab = pd.crosstab(df['education'], df['industry'], values=df['salary'], aggfunc='mean', margins=True, margins_name='Average')
plt.figure(figsize=(10, 8))
sns.heatmap(average_salary_crosstab, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=.5)
plt.title('Heatmap of Average Salary Crosstab')
plt.show()
```
